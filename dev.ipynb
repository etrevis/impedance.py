```python
import impedance
from impedance.models.circuits import CustomCircuit
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
tf_test = CustomCircuit(initial_guess=[2,30,10**-14,10**2,8*10**-1,1*10**1, 10**1, 1],
                              circuit='R_0-Tf')

tf_test_2 = CustomCircuit(initial_guess=[2,30,10**-14,10**2,1*10**0,10**1, 10**1, 0.9],
                              circuit='R_0-Tf')

t_test = CustomCircuit(initial_guess=[2,30,10**-14,10**1,10**1,],
                              circuit='R_0-T')

TLMQ_test = CustomCircuit(initial_guess=[2,30 ,10**1, 1],
                              circuit='R_0-TLMQ')

Zarc_test = CustomCircuit(initial_guess=[2,10 ,10**-3, 0.8],
                              circuit='R_0-Zarc')


freq = np.logspace(-4,6,300)
Z_tf = tf_test.predict(freq)
Z_tf_2 = tf_test_2.predict(freq)
Z_t = t_test.predict(freq)
Z_TLMQ = TLMQ_test.predict(freq)
Z_Zarc = Zarc_test.predict(freq)


fig = go.Figure(data=go.Scatter(x=Z_tf.real, y=-Z_tf.imag, mode='markers', customdata=freq, hovertemplate='<br>Re: %{x:.1f} </br><br>-IM: %{y:.1f} </br> <br>f: %{customdata:.1e} </br>'),)  
fig.add_trace(go.Scatter(x=Z_TLMQ.real, y=-Z_TLMQ.imag, mode='markers', customdata=freq, hovertemplate='<br>Re: %{x:.1f} </br><br> -IM: %{y:.1f} </br><br> f: %{customdata:.1e} </br>'))
fig.add_trace(go.Scatter(x=Z_tf_2.real, y=-Z_tf_2.imag, mode='markers', customdata=freq, hovertemplate='<br>Re: %{x:.1f} </br><br> -IM: %{y:.1f} </br><br> f: %{customdata:.1e} </br>'))
fig.add_trace(go.Scatter(x=Z_t.real, y=-Z_t.imag, mode='markers', customdata=freq, hovertemplate='<br>Re: %{x:.1f} </br><br> -IM: %{y:.1f} </br> <br>f: %{customdata:.1e} </br>')) 
fig.update_yaxes(
    scaleanchor = "x",
    scaleratio = 1, )
fig.update_layout(
    autosize=True,
    width=800,
    height=1200,
    margin=dict(l=10, r=10, t=10, b=10),
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01))
fig.show()
```
