```python
import impedance
from impedance.models.circuits import CustomCircuit
import numpy as np
import plotly.express as px
```

```python
test = CustomCircuit(initial_guess=[2,100,10**-14,10**2,10**1,10**2, 10**0],
                              circuit='R_0-Tf')


freq = np.logspace(-4,6,300)
Z = test.predict(freq)
fig = px.scatter(x=Z.real, y=-Z.imag)
fig.update_yaxes(
    scaleanchor = "x",
    scaleratio = 1, )
fig.show()
```
